```python
from __future__ import absolute_import, division, print_function,unicode_literals
import scipy.stats as stats
import functools
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option("display.precision", 2)
```

```python
white_wine_data = pd.read_csv('./dataset/winequality-white.csv', sep=';')
white_wine_data['type'] = '0'
red_wine_data = pd.read_csv('./dataset/winequality-red.csv', sep=';')
red_wine_data['type'] = '1'
all_data = white_wine_data.append(red_wine_data)
```

```python
data = all_data.describe()['quality']
print(data['min'], data['max'], data['mean'], data['std'])
x = np.linspace(data['min'], data['max'], 1000)
plt.figure(figsize=(20, 10))
ax = all_data['quality'].hist( bins=6)
# plt.plot(x, stats.norm.pdf(x, data['mean'], data['std']), linewidth=4)
plt.ylabel('Norm. Num. of records')
plt.xlabel('Wine Quality')
```

```python
data = white_wine_data.describe()['quality']
print(data['min'], data['max'], data['mean'], data['std'])
x = np.linspace(data['min'], data['max'], 1000)
plt.figure(figsize=(20, 10))
ax = white_wine_data['quality'].hist(density=True, bins=6)
# plt.plot(x, stats.norm.pdf(x, data['mean'], data['std']), linewidth=4)
plt.ylabel('Norm. Num. of records')
plt.xlabel('Wine Quality')
```

```python
data = red_wine_data.describe()['quality']
print(data['min'], data['max'], data['mean'], data['std'])
x = np.linspace(data['min'], data['max'], 1000)
plt.figure(figsize=(20, 10))
ax = red_wine_data['quality'].hist(density=True, bins=6)
plt.plot(x, stats.norm.pdf(x, data['mean'], data['std']), linewidth=4)
plt.ylabel('Norm. Num. of records')
plt.xlabel('Wine Quality')
```
